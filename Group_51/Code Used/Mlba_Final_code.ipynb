```python
from sklearn import preprocessing
label = preprocessing.LabelEncoder()

lbl = label.fit_transform(train_data['Labels'])
```

```python
train, test = train_test_split(train_data, test_size=0.2, random_state=1)
X_train = train[train.columns[1:129]]
y_train = train['Labels']
X_test = test[test.columns[1:129]]
y_test = test['Labels']
```

```python
#X_new_test = test[test_data.columns[2:21]]
#y_new_test = test['Label']
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(X_train)

X_train = scale.transform(X_train)
X_test = scale.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(activation='relu', learning_rate='constant', solver='adam', hidden_layer_sizes=(870, 720, 630, 450, 360), max_iter = 1000, random_state=1)
mlp.fit(X_train, y_train)
```

```python
predict = mlp.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, predict))
print(classification_report(y_test, predict))
```
