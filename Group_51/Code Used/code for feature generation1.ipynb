Variance threshold 

```python
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(1.0) # This will remove all zero-variance features
sf = sel.fit_transform(X) # The file will be saved in sf
sc = X.columns[sel.get_support(indices=True)] #This saves the columns names in sc variable
X[sc].head()
RES = X[sc]
RES.to_csv('ANS.csv',index = False)
```

model information gain


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
fs = SelectKBest(score_func=mutual_info_classif, k=50) # This will select top 50 features using mutual information gain
sf = fs.fit_transform(X,y) # It learns relationship and transfrom the data
sc = X.columns[fs.get_support(indices=True)] #This saves the columns names in sc variable
X[sc].head() #This will show the first few rows of selected features

```

ANOVA 


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
fs = SelectKBest(score_func=f_classif, k=50) # This will select top 50 features using ANOVA F-value
sf = fs.fit_transform(X,y) # It learns relationship and transfrom the data
sc = X.columns[fs.get_support(indices=True)] #This saves the columns names in sc variable
X[sc].head() #This will show the first few rows of selected features
```

WRAPPER methods


RFE


```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
rfe = RFE(lr) #It assigns weights to features using logistic regression
sf = rfe.fit(X, y)  # It learns relationship and transfrom the data
sc = X.columns[rfe.get_support(indices=True)] #This saves the columns names in sc variable
X[sc].head() #This will show the first few rows of selected features
X[sc].to_csv('result.csv',header=False,index=False)
```

SFS


```python
import joblib

sys.modules['sklearn.externals.joblib'] = joblib
from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
sfs = SequentialFeatureSelector(knn)#It assigns weights to features using KNN
sf = sfs.fit_transform(X, y) # It learns relationship and transfrom the data
sc = list(sfs.k_feature_names_) #This saves the feature names in the list variable sc
train_data[sc].head()
```
